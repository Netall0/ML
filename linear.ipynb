```python
import pandas as pd
import matplotlib.pyplot as plt


from sklearn.linear_model import LinearRegression
```

# Data

```python
data = pd.read_csv('cost-revenue-clean.csv')
data.sample(n=1000,random_state=42)


data.describe(include='all')
```

```python

plt.figure(figsize=(10,6))
plt.ylabel('кассовый сбор')
plt.xlabel('бюджет')
plt.scatter(data.production_budget_usd,data.worldwide_gross_usd,alpha=0.4)
```

# Model

```python
linReg = LinearRegression()



X = pd.DataFrame(data.production_budget_usd)
y = pd.DataFrame(data.worldwide_gross_usd)


linReg.fit(X,y)
```

```python
linReg.coef_


```

```python
linReg.intercept_
```

```python

plt.figure(figsize=(10,6))
plt.ylabel('кассовый сбор')
plt.xlabel('бюджет')

plt.plot(X,linReg.predict(X),color = "red")
plt.scatter(data.production_budget_usd,data.worldwide_gross_usd,alpha=0.4)
```

```python
linReg.score(X,y)
```
